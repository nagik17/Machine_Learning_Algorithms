```python
def draw_line(coef,intercept, mi, ma):
    # for the separating hyper plane ax+by+c=0, the weights are [a, b] and the intercept is c
    # to draw the hyper plane we are creating two points
    # 1. ((b*min-c)/a, min) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the minimum value of y
    # 2. ((b*max-c)/a, max) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the maximum value of y
    points=np.array([[((-coef[1]*mi - intercept)/coef[0]), mi],[((-coef[1]*ma - intercept)/coef[0]), ma]])
    plt.plot(points[:,0], points[:,1])
```

### Implementing Task 1: Applying SVM

```python
# first plotting all the scatter plots with different regularization strengths. 
ratios = [(100,2), (100, 20),(100, 40), (100, 80)]
plt.figure(figsize=(30,10))
C_arr=[0.001,1,100]
D=[];S=[];ran=1;
for j,i in enumerate(ratios):        
        X_p=np.random.normal(0,0.05,size=(i[0],2)) #choosing random points for data
        X_n=np.random.normal(0.13,0.02,size=(i[1],2))
        y_p=np.array([1]*i[0]).reshape(-1,1)
        y_n=np.array([0]*i[1]).reshape(-1,1)
        X=np.vstack((X_p,X_n)) # stacking positive and negative points
        y=np.vstack((y_p,y_n)) 
        #D.append([X_p,Xn])
        for c in range(len(C_arr)): # code for every subplot 
            plt.subplot(4, 3, ran)
            plt.scatter(X_p[:,0],X_p[:,1])
            plt.scatter(X_n[:,0],X_n[:,1],color='red') 
            clf = SVC(C=C_arr[c],kernel='linear')
            clf.fit(X,y)
            a,b=clf.coef_[0][0],clf.coef_[0][1]
            draw_line([a,b],clf.intercept_,min(y),max(y))
            sv_indices = clf.support_
            for k in list(clf.support_):
                plt.scatter(X[k][0],X[k][1],color ='k') # black points are support vectors
            ran=ran+1; # for subplot number 
plt.xlim(-0.15,0.15) # setting limits to x,y axes 
plt.ylim(0,0.15)
plt.show()
```

### Task 2: Applying Logistic Regression

```python
ratios = [(100,2), (100, 20),(100, 40), (100, 80)]
plt.figure(figsize=(30,10))
C_arr=[0.001,1,100]
D=[];S=[];ran=1;
for j,i in enumerate(ratios):        
        X_p=np.random.normal(0,0.05,size=(i[0],2))
        X_n=np.random.normal(0.13,0.02,size=(i[1],2)) # choosing random points for data
        y_p=np.array([1]*i[0]).reshape(-1,1)
        y_n=np.array([0]*i[1]).reshape(-1,1)
        X=np.vstack((X_p,X_n)) 
        y=np.vstack((y_p,y_n)) 
        #D.append([X_p,Xn])
        for c in range(len(C_arr)):
            plt.subplot(4, 3, ran)
            plt.scatter(X_p[:,0],X_p[:,1])
            plt.scatter(X_n[:,0],X_n[:,1],color='red')
            clf = LogisticRegression(C=C_arr[c])
            clf.fit(X,y) # training data 
            a,b=clf.coef_[0][0],clf.coef_[0][1] # obtaining coefficients
            draw_line([a,b],clf.intercept_,min(y),max(y)) # decision boundary
            ran=ran+1; # for subplot number 
plt.xlim(-0.15,0.15)
plt.ylim(0,0.15)
plt.show()
```
